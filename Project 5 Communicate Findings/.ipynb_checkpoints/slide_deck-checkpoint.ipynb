## Investigation Overview

In this project, I wanted to see the riders' habits when they use the Ford GoBike system in San Francisco I am interested in how different  user groups will contrast each other in features such as day, time, duration and others. 

## Dataset Overview

The dataset is given by Udacity and contains all rides' information for the month of February in 2019 in San Francisco. The dataset consists of 183412 rows and 16 columns initially. 

```python
# import all packages and set plots to be embedded inline
import pandas as pd
import numpy as np
import datetime as dt
from matplotlib import pyplot as plt
import seaborn as sb
import matplotlib as mpl


%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('201902-fordgobike-tripdata.csv')

#create copy of original dataset
bikes = df.copy()

#change to datetime format
bikes['start_time'] = pd.to_datetime(bikes['start_time'])
bikes['end_time'] = pd.to_datetime(bikes['end_time'])

#extract date, hour, day of week, and day from start_time
bikes['start_date'] = bikes['start_time'].dt.strftime('%Y-%m-%d')
bikes['start_hour'] = bikes['start_time'].dt.hour
bikes['start_day_of_week'] = bikes['start_time'].dt.strftime('%A')
bikes['start_day'] = bikes['start_time'].dt.day.astype(int)

#change id's to string
bikes['start_station_id'] = bikes['start_station_id'].astype('str')
bikes['end_station_id'] = bikes['end_station_id'].astype('str')

#change to categorical
bikes['user_type'] = bikes['user_type'].astype('category')
bikes['member_gender'] = bikes['member_gender'].astype('category')
bikes['bike_share_for_all_trip'] = bikes['bike_share_for_all_trip'].astype('category')

#create new column of duration in minutes
bikes['duration_min'] = bikes['duration_sec']/60

#create new column for age of members
bikes['member_age'] = 2019 - bikes['member_birth_year']

#95% of users are 55 years old and below, so thats where I will set the max
bikes['member_age'].describe(percentiles = [ .95])

#Select only the dataset where member_age is 55 or below
bikes = bikes[bikes['member_age']<=55]

#create age groups
bikes['member_age_groups'] = bikes['member_age'].apply(lambda x: 
                                                    '1 - 15' if 1<x<=15.0
                                                  else '15 - 25 years old' if 15.1<x<=25.0
                                                  else '25 - 35 years old' if 25.1<x<=35.0
                                                  else '35 - 45 years old' if 35.1<x<=45.0
                                                  else '45 - 55 years old' if 45.1<x<=55.0
                                                  else x)

#95% of the rides are 25 minutes long at most, so thats where I will set the max
bikes['duration_min'].describe(percentiles = [ .95])

#Select only the dataset where duration is 25.3 minutes and below
bikes = bikes[bikes['duration_min']<=25.3]
```

## Of all riders, what are the proportions between age groups?

50.1% are 25-35 years old, 22.4% are 35-45 years old, 18.2% are 15-25 years old, and 9.3% are 44-55 years. Age 25 to 45 make up to almost 3/4 of all the riders. 

```python
#age groups

age_groups = bikes['member_age_groups'].value_counts()
plt.figure(figsize=[20,15])
uv4 = plt.pie(age_groups, labels=age_groups.index, autopct='%1.1f%%')

mpl.rcParams['font.size'] = 18.0

plt.title('Proportional difference between age groups', size =30)
plt.show()
```

## What is the duration difference for rides between subscribers and customers?

There are many more subscribers in the 2 to 10 min duration than there are of customers. Subscribers are alot less likely to ride more than 12 minutes. Customers also ride more in the 2 to 10 minute range, but does not differ as much as the subscribers. 

```python
#duration and user type
plt.figure(figsize=[20,15])
bv2 = sb.violinplot(data=bikes, x = bikes['user_type'], y = bikes['duration_min'])

plt.xticks(size =15)
plt.yticks(size =15)

plt.xlabel('user type', size=20)
plt.ylabel('duration in minutes', size=20)

plt.title('Duration of rides difference between user types', size =30)
plt.show()
```

## Do Age groups differ in their duration throughout the days of the week?

For the age group 35-45 and 25-35, they are the most consistent in their duration for each day of the week, mostly ranging from 9 to 9.5 minutes. 35-45 have a peak average on Sunday, where it is above 9.5 and a minimum average usage on Monday, below 9 minutes. 25-35 have a peak average on Saturdays with above 0.5 while all other days are about equal. 15-25 have the shortest duration on each day of the week with minimum average of 8.5 minutes on Wednesday and Tuesday, and a peak of nearly 9 minutes on Saturday. THe 45-55 age group has a peak in Saturdays with an average duration of almost 10 minutes, witha 2nd peak on Sunday with an average of 9.5 minutes while other days are mostly the same.

```python
fig = plt.figure(figsize = [20,15])
sb.set(font_scale = 2)
mv5 = sb.pointplot(data = bikes, x = 'start_day_of_week', y = 'duration_min', hue = 'member_age_groups',
           palette = 'tab10', linestyles = '', dodge = 0.5)


plt.show()
```
