# Clean

```python
#Create copies of the 3 tables

archive = df_twitter_archive.copy()

predictions = df_image_predictions.copy()

tweets = tweets_df.copy()
```

### Define Quality Issue #1: drop all rows that are retweets

### Clean Quality Issue #1: 

```python
#drop all rows that are retweets
archive = archive[archive['retweeted_status_id'].isnull()]
```

### Test Quality Issue #1:

```python
#retweet columns now have 0 non null columns, deleting 181 rows
archive.info()
```

### Define Quality Issue #2: drop all rows without dog images

### Clean Quality Issue #2:

```python
#drop all rows without dog images
archive = archive[archive['expanded_urls'].notnull()]
```

### Test Quality Issue #2:

```python
#rows have dropped from 2175 to 2117, deleting 58 rows
archive.info()
```

### Define Quality Issue #3: drop all retweet columns: retweeted_status_id, retweeted_status_user_id, retweeted_status_timestamp

### Clean Quality Issue #3:

```python
#drop retweet columns 
archive.drop(['retweeted_status_id',
                'retweeted_status_user_id',
                'retweeted_status_timestamp'], axis=1, inplace=True)
```

### Test Quality Issue #3:

```python
#retweeted_status_id, retweeted_status_user_id, retweeted_status_timestamp has been dropped
archive.info()
```

### Define Quality Issue #4: drop all reply columns: in_reply_to_status_id, in_reply_to_user_id

### Clean Quality Issue #4:

```python
#drop reply columns
archive.drop(['in_reply_to_status_id',
                'in_reply_to_user_id'], axis=1, inplace=True)
```

### Test Quality Issue #4:

```python
#in_reply_to_status_id, in_reply_to_user_id has been dropped
archive.info()
```

### Define Tidiness Issue #1: Combine dog stage columns into, if there are multiple dog stage in one tweet, combine both stages into one as its own category

### Clean Tidiness Issue #1:

```python
#See which tweets have multiple dog stages with it
archive.loc[(archive[['doggo', 'floofer', 'pupper', 'puppo']] != 'None'
                 ).sum(axis=1) > 1]
```

```python
#Change data type of columns 'doggo', 'floofer', 'pupper', 'puppo' then replace "none" with empty space 
archive['doggo'].replace('None','',inplace=True)
archive['floofer'].replace('None','',inplace=True)
archive['pupper'].replace('None','',inplace=True)
archive['puppo'].replace('None','',inplace=True)
```

```python
#combine 'doggo', 'floofer', 'pupper', 'puppo' into one column: dog_stage
archive['dog_stage'] = archive['doggo'] + archive['floofer'] + archive['pupper'] + archive['puppo']
```

```python
#view dog_stage column to see which categories need commas
archive['dog_stage'].value_counts()
```

```python
#add comma to any rows with 2 dog stages, and change empty string into np.nan

archive['dog_stage'].replace({'doggopupper':'doggo, pupper', 
                              'doggofloofer':'doggo, floofer',
                             'doggopuppo':'doggo, puppo',
                             '':np.nan}, inplace=True)
```

### Test Tidiness Issue #1

```python
#Dog stage column has all correct dog stage labels
archive['dog_stage'].value_counts()
```

### Define Quality Issue #5: drop doggo, floofer, pupper, and puppo columns

### Clean Quality Issue #5: 

```python
#drop doggo, floofer, pupper, and puppo columns
archive.drop(['doggo',
                'floofer',
                'pupper',
                'puppo'], axis=1, inplace=True)
```

### Test Quality Issue #5: 

```python
#doggo, floofer, pupper, and puppo columns has been dropped
archive.info()
```

### Define Quality Issue #6: archive's "timestamp" column should be in datetime format

### Clean Quality Issue #6:

```python
#change timestamp to datetime format
archive['timestamp'] = archive['timestamp'].astype('datetime64[ns]')
```

### Test Quality Issue #6:

```python
#timestamp column is not datetime format
archive.info()
```

### Define Quality Issue #7: archive's "source" should be cleaned and in text format

### Clean Quality Issue #7:

```python
#change source url to text 

def label_source(x):
    if x['source'] == '<a href="http://twitter.com/download/iphone" rel="nofollow">Twitter for iPhone</a>':
        return 'Twitter for iPhone'
    elif x['source'] == '<a href="http://vine.co" rel="nofollow">Vine - Make a Scene</a>':
        return 'Vine - Make a Scene'
    elif x['source'] == '<a href="http://twitter.com" rel="nofollow">Twitter Web Client</a>':
        return 'Twitter Web Client'
    elif x['source'] == '<a href="https://about.twitter.com/products/tweetdeck" rel="nofollow">TweetDeck</a>':
        return 'TweetDeck'
    else:
        return "unlabeled"
    return

archive['source'] = archive.apply(lambda x: label_source(x), axis = 1)
```

### Test Quality Issue #7: archive's source has been cleaned and is in string format

```python
archive['source'].value_counts()
```

### Define Quality Issue #8: archive's "rating_numerator" should be type float

### Clean Quality Issue #8: 

```python
archive['rating_numerator'] = archive['rating_numerator'].astype(float)
```

### Test Quality Issue #8:

```python
#archive's rating_numerator is now float
archive.info()
```

### Define Tidiness Issue #2: Combine archive, image_predictions, and tweet_json

### Clean Tidiness Issue #2: 

```python
#combine archive, predictions, and tweet_df on tweet_id
twitter_archive_master = archive.merge(predictions, on='tweet_id').merge(tweets, on='tweet_id')
```

### Test Tidiness Issue #2:

```python
#All columns from the 3 datasets are present, with tweet_id combined as the primary key
twitter_archive_master.info()
```

### Visualization #1: Which dog breeds are the most tweeted about?

```python
#select only dog columns with high confidence predictions in p1 predictions, and also whose name is captialized
get_dogs = vis[vis['p1_conf'] > 0.80] 
correct_dogs = get_dogs[get_dogs['p1'].str[0].str.isupper()]
```

```python
#Visualize top ten dogs with the highest count of tweets
correct_dogs['p1'].value_counts()[:10].plot(kind = 'bar', figsize=(10, 10))
```

### Visualization #1 Answer: The most popular to tweet and is correctly predicted breeds of dogs are the Prembroke as 1st and Labrador Retriever as 2nd, while Samoyed and Pomeranian are about tied for 3rd. 

### Insight #1: What are the top 10 breeds for favorite count?

```python
#Select correct_dog's p1 and favorite count
select_dogs = correct_dogs[['p1', 'favorite_count']]
fav_dogs = select_dogs.sort_values('favorite_count', ascending = False).groupby('p1').head(10)
fav_dogs.head(10)
```

### Insight #1 Answer: The labrador retriever is in 1st, 3rd, 8th and 9th place of most favorite count. While the Prembroke is in 5th and 7th place. French bull dog is in 2nd and 9th place. While the Chesapeake Bay Retriever is in 4th and the Samoyed is in 6th. The top 10 are dominated by a few breeds. 

### Insight #2: Which hours of the day are tweets most frequently created?

```python
#Get value_counts of the master dataset's Hour column

hours = archive['timestamp'].dt.hour.astype(str) + " o'clock"
hours.value_counts()
```

### Insight #2 Answer: 1 am is when most people are tweeting and rating dogs, followed by 12 am, and 2 am. 4 pm is ranked 4th and 5 pm is ranked 6th, these are the only top hours when people are tweeting about dogs not in the evening time. 

### Insight #3: What are the top 10 dog breeds for highest rating?

```python
#Get the dog breeds and its ratings from correct_dogs
dog_ratings = correct_dogs[['p1', 'rating_numerator']]
dog_ratings.groupby('p1').agg({'rating_numerator': 'mean'}).sort_values(by='rating_numerator', ascending =False).head(10)
```

### Insight #3 Answer: The highest ranked dog breed is the West Highland Whitte Terrier, with a rating that is almost double that of the 2nd place ranked. Ranks 2nd through 10th is relatively close together when compared with the distance between 1st and 2nd.
